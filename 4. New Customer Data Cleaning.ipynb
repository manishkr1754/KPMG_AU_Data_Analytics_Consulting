```python
# drop the unnamed columns with unknown datapoints
del new_customer['Unnamed: 16']
del new_customer['Unnamed: 17']
del new_customer['Unnamed: 18']
del new_customer['Unnamed: 19']
del new_customer['Unnamed: 20']
```

```python
new_customer.info()
```

```python
# filling in the missing value for last_name
new_customer['last_name'].fillna(value = 'Missing', inplace=True)
```

```python
# check the entered data
new_customer['address'].nunique()
```

```python
# make sure all dates are in the correct format
new_customer['DOB'] = new_customer['DOB'].dt.date
new_customer['DOB']
```

```python
# check the frequency of different job titles
new_customer['job_title'].value_counts()
```

```python
new_customer['job_title'].mode()[0]
```

```python
new_customer['job_title'].fillna(new_customer['job_title'].mode()[0], inplace=True)
```

```python
# check the frequency of job industries
new_customer['job_industry_category'].value_counts()
```

```python
new_customer['job_industry_category'].mode()[0]
```

```python
new_customer['job_industry_category'].fillna(new_customer['job_industry_category'].mode()[0], inplace=True)
```

```python
# make sure that the column names are in the correct format
new_customer.columns = map(str.lower, new_customer.columns)
new_customer.columns = map(str.strip, new_customer.columns)
```

```python
# round the values to the correct format
new_customer['value'] = new_customer['value'].round(3)
new_customer
```
